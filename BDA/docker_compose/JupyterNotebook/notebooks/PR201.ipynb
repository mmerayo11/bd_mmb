```python
import redis

r = redis.Redis(
    host='redis',
    port=6379,
    db=0,
    decode_responses=True
)
```

Pasos 1 y 2.

```python
r.set("app:version", "1.0")
r.get("app:version")
```

Paso 3

```python
r.set("app:version", "1.1")
```

Paso 4, 5 y 6

```python
r.set("contador:descargas", 0)
r.incrby ("contador:descargas", 5)
r.decrby ("contador:descargas", 2)
r.get ("contador:descargas")
```

Pasos 7, 8, 9 y 10 

```python
r.set("app:estado", "activo")
r.set("app:estado", "mantenimiento")
r.set ("mensaje:bienvenida", "Hola alumno")
r.expire("app:estado", 30)
```

Pasos 11 y 12

```python
print(r.exists("app:estado"))
r.delete("app:version")
if not r.exists("app:version"):
    print("app:version ha sido eliminada con éxito")
```
